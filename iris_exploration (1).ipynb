# Task 1: Exploring and Visualizing the Iris Dataset
**Objective:** Load, inspect, and visualize the Iris dataset using pandas, matplotlib, and seaborn.

## Step 1: Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean visual style for all plots
sns.set_theme(style='whitegrid')

print('Libraries imported successfully!')
```

## Step 2: Load the Dataset
We load the Iris dataset directly from seaborn's built-in datasets.

```python
# Load Iris dataset from seaborn
df = sns.load_dataset('iris')

print('Dataset loaded successfully!')
print(f'Type: {type(df)}')
```

## Step 3: Inspect Dataset Structure
Use `.shape`, `.columns`, and `.head()` to understand the data.

```python
# Shape: (rows, columns)
print('Shape (rows, columns):', df.shape)

# Column names
print('\nColumns:', df.columns.tolist())

# First 5 rows
print('\nFirst 5 rows:')
df.head()
```

## Step 4: Basic Summary Statistics
Use `.info()` and `.describe()` to understand data types and statistics.

```python
# Data types and null check
print('Dataset Info:')
df.info()
```

```python
# Statistical summary for numeric columns
print('Statistical Summary:')
df.describe()
```

```python
# Count of each species class
print('Species value counts:')
print(df['species'].value_counts())
```

## Step 5: Scatter Plot — Relationship Between Variables
A scatter plot shows how two numeric variables relate to each other, colored by species.

```python
plt.figure(figsize=(8, 5))

sns.scatterplot(
    data=df,
    x='sepal_length',
    y='petal_length',
    hue='species',       # color by species
    palette='Set1',
    s=80,                # dot size
    alpha=0.8
)

plt.title('Scatter Plot: Sepal Length vs Petal Length', fontsize=14)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Species')
plt.tight_layout()
plt.show()

# Observation: Setosa is clearly separated; Versicolor and Virginica overlap slightly.
```

## Step 6: Histogram — Distribution of Each Feature
Histograms show how values of a feature are spread (distribution shape).

```python
# Plot histograms for all 4 numeric features in a 2x2 grid
numeric_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

fig, axes = plt.subplots(2, 2, figsize=(10, 7))
axes = axes.flatten()  # flatten to easily loop through

for i, col in enumerate(numeric_cols):
    axes[i].hist(df[col], bins=20, color='steelblue', edgecolor='white', alpha=0.85)
    axes[i].set_title(f'Distribution of {col.replace("_", " ").title()}')
    axes[i].set_xlabel(col.replace('_', ' ').title() + ' (cm)')
    axes[i].set_ylabel('Frequency')

plt.suptitle('Histograms of Iris Features', fontsize=15, y=1.02)
plt.tight_layout()
plt.show()

# Observation: Petal length and width show bimodal distributions — two humps —
# suggesting clear separation between setosa and the other two species.
```

## Step 7: Box Plot — Spread and Outliers Per Species
Box plots show the median, quartiles, and outliers for each group.

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 7))
axes = axes.flatten()

for i, col in enumerate(numeric_cols):
    sns.boxplot(
        data=df,
        x='species',
        y=col,
        palette='Set2',
        ax=axes[i]
    )
    axes[i].set_title(f'Box Plot: {col.replace("_", " ").title()} by Species')
    axes[i].set_xlabel('Species')
    axes[i].set_ylabel(col.replace('_', ' ').title() + ' (cm)')

plt.suptitle('Box Plots of Iris Features by Species', fontsize=15, y=1.02)
plt.tight_layout()
plt.show()

# Observation: Setosa has the smallest petal dimensions with minimal spread.
# Virginica has the largest values. A few outliers are visible in sepal_width.
```

## Step 8 (Bonus): Pairplot — All Relationships at Once
A pairplot shows scatter plots for every pair of features, with histograms on the diagonal.

```python
sns.pairplot(df, hue='species', palette='Set1', diag_kind='kde', height=2.5)
plt.suptitle('Pairplot of Iris Dataset', y=1.02, fontsize=14)
plt.show()

# Observation: Petal features (length & width) are the most useful for separating species.
# Setosa is always clearly separated. Versicolor and Virginica overlap in some features.
```
